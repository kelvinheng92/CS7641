```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit
from IPython.display import display
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

### Load dataset

```python
df = pd.read_csv('dataset1/data.csv')
display(df.shape)
display(df.head())
```

### Print statistics of dataset

```python
df.describe()
```

### Keep only Dirtiness label

```python
df.drop(['Pastry', 'Z_Scratch', 'K_Scatch', 'Stains', 'Bumps', 'Other_Faults'], axis=1, inplace=True)
df.rename(columns={'Dirtiness': 'label'}, inplace=True)
label = ['label']
features = list(set(df.columns) - set(label))
display(df.shape)
display(df.head())
```

### MinMax scaler to normalize the feature values

```python
X = df[features].values
scaler = MinMaxScaler()
df[features] = scaler.fit_transform(df[features])
display(df.shape)
display(df.head())
```

```python
temp_label = df.label.copy()
temp_label[temp_label == 0] = 'no fault'
temp_label[temp_label == 1] = 'dirty'
sns.countplot(temp_label)
```

```python
X = df[features]
Y = df[label]
from sklearn.model_selection import StratifiedShuffleSplit
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=1)
sss.get_n_splits(X, Y)
for train_index, test_index in sss.split(X, Y):
    X_trainval, X_test = X.iloc[train_index], X.iloc[test_index]
    Y_trainval, Y_test = Y.iloc[train_index], Y.iloc[test_index]
    trainval_df = pd.concat([X_trainval, Y_trainval], axis=1)
    test_df = pd.concat([X_test, Y_test], axis=1)
```

#### Output processed dataframe to a csv

```python
trainval_df.to_csv('dataset1/processed_trainval_data.csv', index=False)
test_df.to_csv('dataset1/processed_test_data.csv', index=False)
```
